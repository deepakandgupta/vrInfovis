4. Stacked bar chart for Shadows and Lightmapping
A stacked bar chart can be created to visualize the frequency distribution of the shadows and lightmapping settings for each type of light.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv("Lighting_simulation.csv")

# Group the data by Type, Shadows, and Lightmapping, and count the frequency of each group
grouped_df = df.groupby(["Type", "Shadows", "Lightmapping"]).size().reset_index(name="count")

# Pivot the grouped data to create a multi-level index, with Type as the first level and Shadows/Lightmapping as sub-levels
pivoted_df = grouped_df.pivot(index="Type", columns=["Shadows", "Lightmapping"], values="count")

# Create a stacked bar chart to visualize the frequency distribution of Shadows and Lightmapping for each Type of light
pivoted_df.plot(kind="bar", stacked=True)
plt.title("Shadows and Lightmapping for each Type of Light")
plt.xlabel("Type of Light")
plt.ylabel("Frequency")
plt.show()
```

This code will create a stacked bar chart with the x-axis representing the different types of lights (Point Light, Directional Light, Spot Light, and Area Light), and the y-axis representing the frequency of each combination of Shadows and Lightmapping settings. The bars for each type of light will be stacked, with each color representing a different combination of Shadows and Lightmapping settings. The stacked bar chart can help to visualize the distribution of Shadows and Lightmapping settings across the different types of lights, and identify any patterns or trends in their usage.

5. Pie chart for Baking
A pie chart can be created to visualize the proportion of lights that have been baked vs. those that have been set to real-time.

6. Heatmap for Color and Location
A heatmap can be created to visualize the distribution of colors across different locations in the dataset.

```python
import pandas as pd
import seaborn as sns

# Load the dataset into a pandas DataFrame
df = pd.read_csv("Lighting_simulation.csv")

# Create a pivot table to calculate the frequency of each color in each location
color_location_df = df.pivot_table(index="Color", columns="Location", values="Type", aggfunc="count")

# Create a heatmap to visualize the distribution of colors across different locations
sns.heatmap(color_location_df, cmap="YlGnBu", annot=True, fmt="d")
```

This code will create a heatmap with colors representing the frequency of each color in each location. The heatmap can help to quickly identify the dominant colors in each location and any patterns or trends in the distribution of colors across different locations. The "cmap" parameter can be used to change the color map of the heatmap. In this example, the "YlGnBu" color map is used, which goes from yellow to green to blue, with higher values being represented by darker shades of blue. The "annot" parameter is set to True to display the frequency values in each cell of the heatmap. The "fmt" parameter is set to "d" to format the frequency values as integers.

```python
# import matplotlib.pyplot as plt
# import pandas as pd

# df = pd.read_csv('Lighting_simulation.csv')

# # assuming your dataset is stored as a pandas DataFrame called 'df'
# x = df['Type']
# y = df['Color']
# sizes = df['Intensity']

# fig, ax = plt.subplots(figsize=(10, 8))
# scatter = ax.scatter(x, y, s=sizes, alpha=0.5)

# # add legend with size markers
# handles, labels = scatter.legend_elements(prop="sizes", alpha=0.5)
# legend = ax.legend(handles, labels, loc="upper right", title="Intensity")

# ax.set_xlabel('Type')
# ax.set_ylabel('Color')
# plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# create a sample dataframe
df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Yellow', 'Red', 'Blue', 'Green', 'Yellow'],
                   'Value': [2, 4, 6, 8, 10, 12, 14, 16]})

# create a dictionary to store the frequency of each color in the dataframe
color_freq = {}
for color in df['Color']:
    if color.lower() in color_freq:
        color_freq[color.lower()] += 1
    else:
        color_freq[color.lower()] = 1

# create a list of colors to be used for the pie chart
colors = ['red', 'blue', 'green', 'yellow']

# create a list of frequencies of each color in the dataframe
freq = [color_freq[color.lower()] if color.lower() in color_freq else 0 for color in colors]

# plot the pie chart
plt.pie(freq, labels=colors, colors=colors, startangle=90, autopct='%1.1f%%')
plt.title('Color Frequencies')

# add legend
plt.legend(title='Colors', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# add circles to the pie chart to make it a donut chart
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# add value labels to the pie chart
for i, color in enumerate(colors):
    angle = (freq[i] / sum(freq)) * 360
    x = np.cos(np.deg2rad(90 - angle / 2))
    y = np.sin(np.deg2rad(90 - angle / 2))
    plt.text(x, y, str(freq[i]), ha='center', va='center', color='white')

plt.show()
```

This code creates a pie chart that shows the frequency of each color in the dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

df = pd.read_csv('Lighting_simulation.csv')

le = LabelEncoder()
df['Type'] = le.fit_transform(df['Type'])

colors = le.fit_transform(df['Color'])
intensities = df['Intensity']
types = df['Type']

cmap = plt.cm.get_cmap('Set3', len(df['Type'].unique()))

markers = ['o', 's', 'd', 'p', '*', 'h', 'v', '^', '>', '<']

for i, t in enumerate(df['Type'].unique()):
    mask = df['Type'] == t
    ax.scatter(types[mask], colors[mask], intensities[mask], 
               c=[cmap(i)], label=t, marker=markers[i], s=100)

ax.set_xlabel('Light Type')
ax.set_ylabel('Color')
ax.set_zlabel('Intensity')

# Create a legend with markers and labels
handles, labels = ax.get_legend_handles_labels()
unique_labels = list(set(labels))
marker_handles = [plt.Line2D([0,0],[0,0],color='black', marker=markers[i], linestyle='') for i in range(len(unique_labels))]
ax.legend(marker_handles, unique_labels)

plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Load the data into a Pandas dataframe
df = pd.read_csv('Lighting_simulation.csv')

# Group the data by location, type, and intensity and sum the intensity values
grouped_df = df.groupby(['Location', 'Type'], as_index=False)['Intensity'].sum()

# Define the scatter plot
fig = px.scatter(grouped_df, x='Type', y='Intensity', color='Location', size='Intensity', size_max=50,
                 hover_data={'Location': True, 'Type': True, 'Intensity': ':.2f'})

# Add dropdown menu for location selection
fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(label="Livingroom",
                     method="update",
                     args=[{"data": [{"x": [list(grouped_df[grouped_df['Location'] == 'Livingroom']['Type'])],
                                       "y": [list(grouped_df[grouped_df['Location'] == 'Livingroom']['Intensity'])]}],
                            "traces": [0]}]),
                dict(label="Kitchen",
                     method="update",
                     args=[{"data": [{"x": [list(grouped_df[grouped_df['Location'] == 'Kitchen']['Type'])],
                                       "y": [list(grouped_df[grouped_df['Location'] == 'Kitchen']['Intensity'])]}],
                            "traces": [0]}]),
                dict(label="Bedroom",
                     method="update",
                     args=[{"data": [{"x": [list(grouped_df[grouped_df['Location'] == 'Bedroom']['Type'])],
                                       "y": [list(grouped_df[grouped_df['Location'] == 'Bedroom']['Intensity'])]}],
                            "traces": [0]}])
            ],
            direction="down",
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        )
    ])

# Show the plot
fig.show()
```
